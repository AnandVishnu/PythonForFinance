```python
# Normal Dist
ds1 = stats.norm(0, 5).rvs(100)

# Two gamma dist. Concatenated together
ds2 = np.concatenate([stats.gamma(5).rvs(50)-1, 
                     -1*stats.gamma(5).rvs(50)])

#box plot
sns.boxplot(data=[ds1, ds2], whis=np.inf)
```
