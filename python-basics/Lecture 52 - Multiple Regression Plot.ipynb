```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot('total_bill', 'tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips,
          scatter_kws={'marker': 'o', 'color': 'indianred'},
          line_kws={'linewidth': 1, 'color': 'blue'})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, order=2,
          scatter_kws={'marker': 'o', 'color': 'indianred'},
          line_kws={'linewidth': 1, 'color': 'blue'})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, fit_reg=False)
```

```python
tips.head()
```

```python
tips['tip_pect']=100*tips['tip']/tips['total_bill']
```

```python
tips.head()
```

```python
sns.lmplot('size', 'tip_pect', tips)
```

```python
sns.lmplot('size', 'tip_pect', tips, x_jitter=0.1)
```

```python
sns.lmplot('size', 'tip_pect', tips, x_estimator=np.mean)
```

```python
sns.lmplot('total_bill', 'tip_pect', tips, hue='sex', markers=['x', 'o'])
```

```python
sns.lmplot('total_bill', 'tip_pect', tips, hue='day')
```

```python
#local regression

sns.lmplot('total_bill', 'tip_pect', tips, lowess=True, line_kws={'color': 'black'})
```

```python
sns.regplot('total_bill', 'tip_pect', tips)
```

```python
fig, (a1, a2) = plt.subplots(1, 2, sharey=True)
sns.reg_plot('total_bill', 'tip_pect', tips, ax=a1)
sns.violinplot(tips['tip_pect'], tips['size'], color='Reds_r')
```
