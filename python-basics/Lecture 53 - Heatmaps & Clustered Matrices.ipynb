```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
flights_df = sns.load_dataset('flights')
```

```python
flights_df.head()
```

```python
fdp = flights_df.pivot('month', 'year', 'passengers')
```

```python
fdp
```

```python
sns.heatmap(fdp)
```

```python
sns.heatmap(fdp, annot=True, fmt='d')
```

```python
sns.heatmap(fdp, annot=True, fmt='d', center=fdp.loc['January', 1955])
```

```python
f, (a1, a2) = plt.subplots(2,1)
yearly_flights = fdp.sum()
years = pd.Series(yearly_flights.index.values)
years = pd.DataFrame(years)

flts = pd.Series(yearly_flights.values)
flts = pd.DataFrame(flts)

y_df = pd.concat((years, flts), axis= 1)
y_df.columns= ['Year', 'Flights']

sns.barplot(x='Year', y='Flights', data=y_df, ax=a1)
sns.heatmap(fdp, cmap='Blues', ax=a2, cbar_kws={'orientation':'horizontal'})
```
