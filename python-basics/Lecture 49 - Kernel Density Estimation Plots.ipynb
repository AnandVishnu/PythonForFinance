```python
import numpy as np
import pandas as pd
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#manual kde construction

d1 = randn(25)
```

```python
sns.rugplot(d1)
plt.ylim(0,1)
```

```python
plt.hist(d1, alpha=0.3)
sns.rugplot(d1)
```

```python
sns.rugplot(d1)

x_min = d1.min() - 2
x_max = d1.max() + 2
x_axi = np.linspace(x_min, x_max, 100)
bandw = ((4*d1.std()**5)/(3*len(d1)))**0.2
klist = []

for dp in d1:
    #create a kernel for each point and append it to the kernel list
    k = stats.norm(dp, bandw).pdf(x_axi)
    klist.append(k)
    
    #scale for plotting
    k = k / k.max()
    k = k * 0.4
    plt.plot(x_axi, k, color='grey', alpha= 0.5)

plt.ylim(0,1)
```

```python
sum_of_kde = np.sum(klist, axis=0)
fig = plt.plot(x_axi, sum_of_kde, color='indianred')
sns.rugplot(d1)
```

```python
sns.kdeplot(d1)
```

```python
sns.rugplot(d1, color='black')

for bw in np.arange(0.5, 2, 0.25):
    sns.kdeplot(d1, bw=bw, lw=1.8, label=bw)
```

```python
ko = ['biw', 'cos', 'epa', 'gau', 'tri', 'triw']
for k in ko:
    sns.kdeplot(d1, kernel=k, label=k)
```

```python
ko = ['biw', 'cos', 'epa', 'gau', 'tri', 'triw']
for k in ko:
    sns.kdeplot(d1, kernel=k, label=k, shade=True)
```

```python
sns.kdeplot(d1, vertical=True)
```

```python
sns.kdeplot(d1, cumulative=True)
```
